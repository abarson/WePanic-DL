```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys, os
```

```python
redscale_qbc = pd.read_csv('../top2/QBC.csv')
greyscale_qbc = pd.read_csv('../../top1_copy/QBC.csv')
combined = pd.DataFrame()

combined['redscale'] = redscale_qbc.QBC_hr
combined['greyscale'] = greyscale_qbc.QBC_hr
combined['actual'] = greyscale_qbc.actual_hr

combined['avg'] = (combined.redscale + combined.greyscale)/2
combined['SE']  = (combined.actual - combined.avg)**2
combined
```
